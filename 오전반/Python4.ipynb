## 멜론 TOP100 리스트 - 가수별 곡수

```python
# 1) 데이터를 로컬에 다운받아, 멜론_TOP100.csv 파일로 저장
import requests

res = requests.get("https://goo.gl/iDUKrL")
with open("멜론_TOP100.csv", "wt", encoding="utf8") as f:
    f.write(res.text)

# 2) 로컬의 CSV 데이터를 리스트(내부 dict)구조로 획득
import csv
with open("멜론_TOP100.csv", "rt", encoding="utf8") as f:
    reader = csv.DictReader(f)
    멜론_top100_리스트 = list(reader)

멜론_top100_리스트
```

```python
len(멜론_top100_리스트)
```

```python
가수별_곡수_사전 = {}  # {'방탄소년단': 1, '아이유': 2}

for 곡 in 멜론_top100_리스트:
    가수 = 곡['가수']
    if 가수 in 가수별_곡수_사전:
        가수별_곡수_사전[가수] += 1
    else:
        가수별_곡수_사전[가수] = 1

가수별_곡수_사전
```

```python
len(가수별_곡수_사전)
```

```python
int()
```

```python
from collections import defaultdict

# 존재하지 않는 Key에 접근할 때,
# KeyError를 발생시키지 않고,
# 해당 Value를 0으로 채워준다.
가수별_곡수_사전 = defaultdict(int)

for 곡 in 멜론_top100_리스트:
    가수 = 곡['가수']
    가수별_곡수_사전[가수] += 1

가수별_곡수_사전
```

```python
# 세번째 풀이
from collections import Counter

가수_리스트 = []

for 곡 in 멜론_top100_리스트:
    가수_리스트.append(곡['가수'])

Counter(가수_리스트)
```

```python
# 네번째

# List Comprehension
가수_리스트 = [
    곡['가수']
    for 곡 in 멜론_top100_리스트]

Counter(가수_리스트)
```

## 반복문에서의 break

```python
# 풀이 1)
for 숫자 in range(2, 10):
    print(숫자)
    for i in range(1, 10):
        print(숫자, i, 숫자 * i)
        if i == 숫자:
            break
```

```python
# 풀이 2)
for 숫자 in range(2, 10):
    print(숫자)
    for i in range(1, 10):
        if i <= 숫자:
            print(숫자, i, 숫자 * i)
```

```python
# 풀이 3)
for 숫자 in range(2, 10):
    print(숫자)
    for i in range(1, 숫자+1):
        print(숫자, i, 숫자 * i)
```

```python
def 구구단_출력(숫자):
    print('--- {}단 ---'.format(숫자))
    for i in range(1, 10):
        print(숫자, i, 숫자 * i)
```

```python
def 부가세_계산(결제금액):
    return 결제금액 // 11
```

```python
def 부가세_계산2(결제금액):
    부가세 = 결제금액 // 11
    부가세_뺀_금액 = 결제금액 - 부가세
    return (부가세_뺀_금액, 부가세)
```

```python
부가세_계산2(20000)
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("http://naver.com")
html = res.text
soup = BeautifulSoup(html, 'html.parser')
tag_list = soup.select('.PM_CL_realtimeKeyword_rolling .ah_k')

message_list = []
for rank, tag in enumerate(tag_list, 1):  # 항상 하위 block 전에 콤마(:)를 씁니다.
    label = tag.text
    message = '{}: {}'.format(rank, label)
    message_list.append(message)
    
메세지 = '\n'.join(message_list)

chat_id = '42478249'  # FIXME: 여러분의 ID
bot.sendMessage(chat_id=chat_id, text=메세지)
```
