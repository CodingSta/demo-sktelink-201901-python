```python
import requests
from bs4 import BeautifulSoup
```

## 서비스 코드 조사

https://www.uplussave.com/dev/lawPpnList.mhp 페이지의 `data-ppnCd` 항목

```python
res = requests.get('https://www.uplussave.com/dev/lawPpnList.mhp')
html = res.text
soup = BeautifulSoup(html, 'html.parser')

for tag in soup.select('[data-ppncd]'):
    label = tag['data-gtm-tracking-label']
    svc_code = tag['data-ppncd']
    print(label, svc_code)
```
