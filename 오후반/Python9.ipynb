### 함수 구성

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlencode
```

```python
def 네이버_웹툰_에피소드_크롤링(title_id, no, prefix=None):
    episode_url = 'https://comic.naver.com/webtoon/detail.nhn'
    params = {
        'titleId': title_id,
        'no': no,
    }
    res = requests.get(episode_url, params=params)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    
    h2_tag = soup.select_one('.detail h2')
    author_tag = h2_tag.select_one('span')
    if author_tag:
        author_tag.extract()  # span 태그 제거
    comic_name = h2_tag.text.strip()
    
    ep_name = soup.select_one('.tit_area h3').text
    
    tag_list = soup.select('.wt_viewer img')
    image_url_list = [tag['src'] for tag in tag_list]  # list comprehension
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
        'Referer': episode_url + '?' + urlencode(params),
    }
    
    for image_url in image_url_list:
        res = requests.get(image_url, headers=headers)
        image_data = res.content
        
        image_name = os.path.basename(image_url)
        image_path = os.path.join('data', comic_name, ep_name, image_name)
        
        if prefix:
            image_path = os.path.join(prefix, image_path)

        dir_path = os.path.dirname(image_path)
        if not os.path.exists(dir_path):
            os.makedirs(dir_path)
        
        print("write to {} ({} bytes)".format(image_path, len(image_data)))
        
        with open(image_path, 'wb') as f:
            f.write(image_data)
```

```python
# 좀비딸 : 715772, no: 24, 23, 22, 21, ...
네이버_웹툰_코드표 = {
    '좀비딸': '715772',
}

네이버_웹툰_에피소드_크롤링(네이버_웹툰_코드표['좀비딸'], '21')#, prefix='c:/웹툰')
```
